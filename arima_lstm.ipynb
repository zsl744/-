```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import r2_score

# Load the data (Ensure your dataframe is named 'df' and contains the 'Frequency' column)
data = df["Frequency"]

# Split data into training and testing sets
train_size = int(len(data) * 0.8)  # Use 80% for training, 20% for testing
train, test = data[:train_size], data[train_size:]

# Fit the ARIMA model with (p=9, d=0, q=9)
model = ARIMA(train, order=(1, 0, 5))
model_fit = model.fit()

# Make predictions
predictions = model_fit.forecast(steps=len(test))

# Calculate R² Score
r2 = r2_score(test, predictions)
print(f"R² Score: {r2:.4f}")

# Plot actual vs predicted values
plt.figure(figsize=(12, 5))
plt.plot(test.index, test, label="Actual", color="blue")
plt.plot(test.index, predictions, label="Predicted", color="red", linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.title("ARIMA (1,0,5) - Actual vs Predicted")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
# Calculate MSE
mse = mean_squared_error(test,predictions)
print(f"Mean Squared Error (MSE): {mse:.4f}")
```
